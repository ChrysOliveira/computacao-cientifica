```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mês
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
todos_produtos = np.array([creme_facial,limpeza_facial,pasta_dentaria,sabonete,shampoo,hidratante])

descricao_meses = np.array(['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez'], dtype=str)
quantidade_meses = np.arange(np.size(creme_facial))

soma_total_meses = [np.sum(todos_produtos[:,x]) for x in np.arange(np.size(quantidade_meses))]


fig, ax = plt.subplots(3,2)
fig.delaxes(ax[2,1])
fig.set_size_inches(20,20)

# Gráfico 1 - Total de produtos Vendidos por mês - Linha
ax[0,0].set_title("Total de produtos vendidos por mes")
ax[0,0].plot(quantidade_meses, soma_total_meses)
ax[0,0].xaxis.set_ticks(quantidade_meses)
ax[0,0].xaxis.set_ticklabels(descricao_meses)
ax[0,0].yaxis.set_label_text("Total de produtos vendidos")

# Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha
ax[0,1].set_title("Total vendido por mes")

ax[0,1].plot(quantidade_meses, creme_facial, label="Creme facial", color="blue")
ax[0,1].plot(quantidade_meses, limpeza_facial, label="Limpeza Facial", color="red")
ax[0,1].plot(quantidade_meses, pasta_dentaria, label="Pasta Dentaria", color="yellow")
ax[0,1].plot(quantidade_meses, sabonete, label="Sabonete", color="green")
ax[0,1].plot(quantidade_meses, shampoo, label="Shampoo", color="orange")
ax[0,1].plot(quantidade_meses, hidratante, label="Hidrante", color="brown")

ax[0,1].xaxis.set_ticks(quantidade_meses)
ax[0,1].xaxis.set_ticklabels(descricao_meses)
ax[0,1].yaxis.set_label_text("Total de produtos vendidos")

ax[0,1].legend()

# Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras
ax[1,0].set_title("Creme facial vs Limpeza Facial")
tamanho_barra = 0.3
ax[1,0].bar(quantidade_meses - tamanho_barra/2, creme_facial, width=tamanho_barra)
ax[1,0].bar(quantidade_meses + tamanho_barra/2, limpeza_facial, width=tamanho_barra)

ax[1,0].xaxis.set_ticks(quantidade_meses)
ax[1,0].xaxis.set_ticklabels(descricao_meses)
ax[1,0].yaxis.set_label_text("Total de produtos vendidos")

# Gráfico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos (1000-1999, 2000-2999, ...)
ax[1,1].set_title("Meses por valores")
ax[1,1].hist(soma_total_meses, edgecolor="white")
ax[1,1].yaxis.set_label_text("Quantidade de meses")
ax[1,1].xaxis.set_label_text("Total de produtos vendidos")


# Gráfico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto
ax[2,0].set_title(r"% de vendas por produto")

total_produtos_vendidos_ano = np.sum(soma_total_meses)
porcentagem_creme_facial =  (np.sum(creme_facial) / total_produtos_vendidos_ano) * 100
porcentagem_limpeza_facial =  (np.sum(limpeza_facial) / total_produtos_vendidos_ano) * 100
porcentagem_pasta_dentaria =  (np.sum(pasta_dentaria) / total_produtos_vendidos_ano) * 100
porcentagem_sabonete =  (np.sum(sabonete) / total_produtos_vendidos_ano) * 100
porcentagem_shampoo =  (np.sum(shampoo) / total_produtos_vendidos_ano) * 100
porcentagem_hidratante =  (np.sum(hidratante) / total_produtos_vendidos_ano) * 100

valores = np.array([porcentagem_creme_facial,porcentagem_limpeza_facial,porcentagem_pasta_dentaria,porcentagem_sabonete,porcentagem_shampoo,porcentagem_hidratante])
produtos = np.array(["Creme Facial","Limpeza Facial","Pasta Dentaria","Sabonete","Shampoo","Hidratante"], dtype=str)

ax[2,0].pie(valores, labels=produtos, autopct="%1.2f%%")

fig.savefig("Graficos.png")

plt.show()
```
