```python
import matplotlib.pyplot as plt
import numpy as np


def taylor_bobo(x):
    return (x-1) - (1/2 * np.power(x - 1, 2)) + (1/3 * np.power(x - 1, 3)) - (1/4 * np.power(x - 1, 4)) + (1/5 * np.power(x - 1, 5))

def taylor(x,n=20):
    taylor = 0.0
    for n in range(1,n):
        numerador = np.power(-1,n+1) * np.power(x-1,n)
        denominador = n
        taylor = np.add(taylor, np.divide(numerador,denominador))
    return taylor

print(f"Calculo bobo em a=2: {taylor_bobo(2)}")
print(f"Calculo otimizado em a=2: {taylor(2)}")
print(f"Calculo numpy em a=2: {np.log(2)}")

pontos_x = np.linspace(1/500, 2.2, 500)
y_bobo = taylor_bobo(pontos_x)
y_otimizado = taylor(pontos_x)
y_numpy = np.log(pontos_x)

fig, ax = plt.subplots()
fig.set_size_inches(20,8)
ax.plot(pontos_x, y_bobo, linewidth=2.0, label="bobo")
ax.plot(pontos_x, y_otimizado, linewidth=2.0, label="otimizado")
ax.plot(pontos_x, y_numpy, linewidth=2.0, label="numpy")
fig.legend()
ax.grid()
```
